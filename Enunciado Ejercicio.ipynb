# CONSTRUCTOR:
1. 1 Inventario (lista de diccionarios): Un atributo para almacenar los productos en el inventario.

Ejemplo de como debería ser: [{'nombre': 'Camisa', 'precio': 20, 'cantidad': 40}, {'nombre': 'Pantalón', 'precio': 30, 'cantidad': 30}]

1. 2 clientes (diccionario): Un atributo para llevar un registro de los clientes de la tienda. 

Cada cliente debe ser representado como un diccionario con las siguientes claves: 'nombre' y 'email'. Al inicio deberá ser un diccionario vacío. 

Además, cada cliente debe tener un historial de compras. Deberá parecerse a:
